```python
import pandas as pd
import sportsreference as sportsreference
from pprint import pprint
```

```python
from sportsreference.nfl.teams import Teams

df=pd.DataFrame()   
teams = []
years = []
for i in range(2002, 2020):
    
    teams = Teams(i)
    for team in teams:
        df = df.append(team.dataframe)
        years.append(i)
    
df["years"]=years
df
```

```python
for col in df.columns:
    print(col)
```
